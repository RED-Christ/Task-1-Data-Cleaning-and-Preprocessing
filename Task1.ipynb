```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

from plotly.offline import iplot, plot
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("netflix_titles/netflix_titles.csv")
df.head(10)
```

Identify and handle missing values using

```python
print(df.isnull().sum())
```

```python
# Replacing missing values with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
```

```python
df.head(10)
```

```python
# Deleting rows where these specific columns are empty
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
```

Remove duplicate rows using .drop_duplicates()

```python
# Count total completely duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")
```

```python
df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df.head(10)
```

```python
unique_values = df['rating'].unique()
print(unique_values)
```

```python
unique_values = df['type'].unique()
print(unique_values)
```

Standardize text values

```python
# Select only text/object columns
text_columns = df.select_dtypes(include=['object']).columns

# Check leading/trailing spaces
for col in text_columns:
    has_spaces = df[col].astype(str).str.match(r'^\s|\s$').any()
    
    if has_spaces:
        print(f"Column '{col}' contains leading/trailing spaces")
        
        # Show examples
        examples = df[df[col].astype(str).str.match(r'^\s|\s$')][col].head(5)
        print("Examples:")
        print(examples.tolist())
        print("-" * 50)
    else:
        print(f"Column '{col}' has NO leading/trailing spaces")
```

```python
df.dtypes
```

```python
# Removes leading and trailing spaces for text columns
text_cols = ['type', 'date_added', 'duration', 'description', 'title', 'director', 'cast', 'country', 'rating', 'listed_in']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
# Select only text/object columns
text_columns = df.select_dtypes(include=['object']).columns

# Check leading/trailing spaces
for col in text_columns:
    has_spaces = df[col].astype(str).str.match(r'^\s|\s$').any()
    
    if has_spaces:
        print(f"Column '{col}' contains leading/trailing spaces")
        
        # Show examples
        examples = df[df[col].astype(str).str.match(r'^\s|\s$')][col].head(5)
        print("Examples:")
        print(examples.tolist())
        print("-" * 50)
    else:
        print(f"Column '{col}' has NO leading/trailing spaces")
```

```python
# Strategy A: Extract only the primary (first) country listed
df['primary_country'] = df['country'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'Unknown')

# Strategy B: Clean genre strings by ensuring uniform spacing around commas
df['listed_in'] = df['listed_in'].apply(lambda x: ", ".join([g.strip().title() for g in x.split(',')]))
```

Convert date formats to a consistent type (e.g., dd-mm-yyyy)

```python
# Convertit la colonne en type datetime standardisé (ex: YYYY-MM-DD)
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
df.head(10)
```

Rename column headers to be clean and uniform (e.g., lowercase, no spaces). 

```python
# Convertit en minuscules, remplace les espaces par des '_' et supprime les espaces aux extrémités
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head(10)
```

```python
df.dtypes
```

```python
# Strategy A: Extract only the primary (first) country listed
df['primary_category'] = df['listed_in'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'Unknown')
df.head(10)
```
